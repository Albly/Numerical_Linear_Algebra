**Task 1.1**

(1 pts) What are the constants $C_1$ and $C_2$ such that $C_1 \|x\|_{1} \leq \|x\|_{\infty} \leq C_2 \| x\|_{1}$ for any vector $x$

Consider both parts of nonequation separately. Assume that $x$ has $N$ dimentions. Also lets write definitions for the given norms:

$$
\|x\|_{1} \stackrel{\mathrm{def}}{=} \sum\limits_{i=1}^{N} |x_i|
\qquad
 \|x\|_{\infty} \stackrel{\mathrm{def}}{=} \max | x_i| 
$$

**Left part:** 

$$
C_1 \|x\|_{1} \leq \|x\|_{\infty}
$$

Due to norm definitions:

$$
C_1 \sum\limits_{i=1}^{N} |x_i| \leq \max | x_i| $$

Let $\max | x_i|  = |x_m|$, then we can rewrite nonequation as follows:
$$
C_1\bigg(  \sum\limits_{i=1, i \neq m}^{N} |x_i| +|x_m|  \bigg) \leq |x_m|
\quad \Rightarrow \quad
C_1 \leq  \frac{|x_m|}{\sum\limits_{i=1, i \neq m}^{N} |x_i| +|x_m|}
$$ 
It's obvious that denumenator cannot be smaller than numenator, and so the fraction have to be $\leq 1 $. Then:
$$
C_1 \leq  \frac{|x_m|}{\sum\limits_{i=1, i \neq m}^{N} |x_i| +|x_m|} \leq1 \Rightarrow C_1 \leq 1
$$ 
However we can have several special cases:

1) When we have only one non zero element:
$$
\sum\limits_{i=1, i \neq m}^{N} |x_i| = 0 \quad \text{and} \quad  |x_m| \neq 0
$$
in such case we again can get that $C_1 \leq 1$

2) When $|x_i| = |x_m|$, $\forall i$ :  
$$
C_1 \leq \frac{x_m}{Nx_m} \Rightarrow C_1 \leq \frac{1}{N}
$$
Finally: 
$$
C_1 \leq 1 \leq \frac{1}{N} \Rightarrow C_1 \leq \frac{1}{N}
$$
**Right part:**

Again we use norm defenitions and the fact that max element is $|x_m|$ : 

$$
\|x\|_{\infty} \leq C_2 \| x\|_{1} 
\quad \Rightarrow \quad 
|x_m| \leq C_2 \bigg(  \sum\limits_{i=1, i \neq m}^{N} |x_i| +|x_m|  \bigg)
$$

$$
\frac{|x_m|}{\sum\limits_{i=1, i \neq m}^{N} |x_i| +|x_m|} \leq C_2
$$
Again, using the fact that denumenator is not less than numenator we can conclude that $1 \leq C_2$
Considering the special case when elements are the same we will get: 
$$
\frac{1}{N} \leq  C_2
$$
Finally: 
$$
C_2 \geq \frac{1}{N} \geq 1 
\quad \Rightarrow \quad  C_2 \geq 1
$$

So:
$$
C_2 \geq 1 \qquad C_1 \leq \frac{1}{N}
$$

**Task 1.2**

(5 pts) Prove that $\| U A \|_F = \| A U \|_F = \| A \|_F$ for any unitary matrix $U$.

Lets write the definition of the Frobenius norm: 
$$
\| A \|_F \stackrel{\mathrm{def}}{=} \sqrt{\sum_{i=1}^n \sum_{j=1}^m |a_{ij}|^2}
$$
It has such property as:
$
\|A\|_F = \sqrt{\text{tr}(A^HA)}
$, where $A^H$ is conjugate transposed matrix $A$:

**First equation**
$$
\|UA\|_F^2 = \text{tr}((UA)^H\times(UA))
$$
Using the fact that $(UA)^H = A^HU^H$ we will get: 
$$
\text{tr}((UA^H)\times(UA)) = \text{tr}(A^HU^HUA) = \text{tr}(A^HA) = \|A\|_F^2
$$
Because $UU^H =  I$ and $A^HIA = A^HA$
$$$$
**second equation:**
$$
\|AU\|_F^2 = \text{tr}((AU)^H\times(AU)) = \text{tr}((AU)\times(AU)^H)
$$ because $\text{tr}(AB) = \text{tr}(BA)$. Then using the fact that $(AU)^H = U^HA^H$:
$$
\text{tr}((AU)\times(AU)^H) = \text{tr}(AUU^HA^H) = \text{tr}(AIA^H) = \text{tr}(AA^H) =  \|A\|_F^2
$$

**Task 1.3**

(5 pts) Prove that $\| U A \|_2 = \| A U \|_2 = \| A \|_2$ for any unitary matrix $U$.

In general matrix norm can be represented as: 
$\|A\| \stackrel{\mathrm{def}}{=} \sup  \left\{{\frac{\|Ax\|}{\|x\|}}: x\neq 0\right\} = \sup \limits _{\|x\|=1} \{\|Ax\|\}$
Spectral norm is represented as follows: 
$
\|A\|_2 \stackrel{\mathrm{def}}{=} \sup \left\{{\frac{\|Ax\|_2}{\|x\|_2}}: x\neq 0\right\} = \sup \limits _{\|x\|_1 = 1} \{\|Ax\|_2 \} = \
\sup \limits _{(x,x) =1}{\sqrt {(Ax,Ax)}}={\sqrt {\lambda _{max}(A^HA)}}
$
Then for our case we can write:
$$
\|UA\|_2 \stackrel{\mathrm{def}}{=}{\sqrt {\lambda _{max}((UA)^HUA)}}
$$
Using the fact that $(AB)^H = B^HA^H$ and $U^HU=I$ we wil get:
$$
\sqrt {\lambda _{max}((UA)^HUA)} = \sqrt {\lambda _{max}(A^HU^HUA)} = \sqrt {\lambda _{max}(A^HIA)} = \sqrt {\lambda _{max}(A^HA)} \stackrel{\mathrm{def}}{=} \|A\|_2 
$$ 
Lets do the same with second equation: 
$$
\|AU\|_2 \stackrel{\mathrm{def}}{=}\sup \limits _{\|x\|_1 = 1} \{\|AUx\|_2 \}
$$
However Unitary matrix doesn't change the norm of vector :$\|Ux\|_2 = \|x\|_2$. Lets denote $\|Ux\|_2 = g$:
$$
\sup \limits _{\|x\|_1 = 1} \{\|AUx\|_2 \} = \sup \limits _{\|g\|_1 = \|x\|_1 = 1} \{\|Ag\|_2 \}\stackrel{\mathrm{def}}{=}\|A\|_2
$$

**Task 2.1**

(5 pts) Using the results from the previous subproblem, prove that $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$. _Hint:_ SVD will help you.

According to the spectral norm definition, it is equal to the largest singular value:
$\|A\|_2 \stackrel{\mathrm{def}}{=} \sqrt{\lambda _{max}(A^{*}A)} = \sigma_1$. 
Lets do singular value decomposition of squared Frobenius norm of matrix $A$:
$$
\|A\|^2_F = \| U \Sigma V^H \|^2_F 
$$
Using unitary invariance property: $\|UAV\|_F = \|A\|_F$ we will get:
$$
\| U \Sigma V^H \|^2_F = \| \Sigma \|_F^2
$$
Now lets use Frobenius norm property: $\|A\|^2_F = \text{tr}(A^HA)$:
$$
\| \Sigma \|_F^2 = \text{tr}(\Sigma^H\Sigma)
$$
But $\Sigma$ is diag matrix. Then $\Sigma^H\Sigma$ is also diag with squared elements:
$$
\text{tr}(\Sigma^H\Sigma) = \sum_{i=1}^{N}{\sigma_i^2}
$$

So, 
$$
\|A\|^2_F = \sum_{i=1}^{N}{\sigma_i^2} \qquad \|A\|_2 = \sigma_1
$$
Also $\text{rank}(A)$ in terms of singuar values is number $K$ of non-zero singular values.


Now, lets prove that $\|A\|_F \leq \sqrt{\text{rank} (A)} \|A\|_2$.
Using our previous results we can rewrite it as follows:
$$
\|A\|_F \leq \sqrt{\text{rank} (A)} \|A\|_2 \quad \Rightarrow \quad \sqrt{\sum_{i=1}^{N}{\sigma_i^2}} \leq \sqrt{K} \sigma_1 \Rightarrow
\sqrt{\sum_{i=1}^{K}{\sigma_i^2} + \sum_{i=K}^{N}{0}} \leq \sqrt{K} \sigma_1
$$
Lets ommit sum of zeros and power this non-equality in 2:
$$
\sqrt{\sum_{i=1}^{K}{\sigma_i^2}} \leq \sqrt{K} \sigma_1 \Rightarrow \sum_{i=1}^{K}{\sigma_i^2} \leq K \sigma^2_1
$$
And this non-equality is always true because sum of $K$ maximum singular value is not less than sum of $K$ all other singular values, which are not more than $\sigma_1$

$$
\det(\exp(S^{-1}AS)) = \det(S^{-1}\exp(A)S)

$$
\det(S^{-1}\exp(A)S) = \det(S^{-1})\det(\exp(A))\det(S) = \frac{1}{\det(S)}\det(A)\det(S) = \det(A)

$$ \Rightarrow \frac{\partial}{\partial V} F(U, V) = D^T U^T U D V - D^T U^T X = D^T U^T (U D V - X)$$

````python
import jax
import jax.numpy as jnp

def generate(a,b):
    '''generates random matrix with size (a,b)'''
    return jax.random.normal(jax.random.PRNGKey(1),(a,b))

@jax.jit
def function(X,U,D,V):
    '''Function which gradient we want to calculate'''
    return 0.5*jnp.linalg.norm(X-U @ D @ V,'fro')**2

@jax.jit
def grad_U(X,U,D,V):
    '''Computes analytical solution of the gradient for @function with respect to U'''
    return (U @ D @ V - X) @ V.T @ D.T

@jax.jit
def grad_V(X,U,D,V):
    '''Computes  solution of the gradient for @function with respect to V'''
    return D.T @ U.T @ ( U @ D @ V - X)

@jax.jit
def autograd_U(X,U,D,V):
    '''Computes automaticaly the gradient for @function with respect to U'''
    g = jax.grad(function, argnums=1)
    return g(X, U, D, V)

@jax.jit
def autograd_V(X,U,D,V):
    '''Computes automaticaly the gradient for @function with respect to V'''
    g = jax.grad(function, argnums=3)
    return g(X, U, D, V)

n,k = 100,200       # matrix sizes
# generate all the matrices that we need
X = generate(n,n)   
U = generate(n,k)
D = jax.numpy.diag(jax.random.normal(jax.random.PRNGKey(1),(1,k))[0,:])
V = generate(k,n)

# Calculate gradient of @function with respect to U 
res_analytical = grad_U(X, U, D, V).block_until_ready()
res_autograd = autograd_U(X, U, D, V).block_until_ready()
print(jnp.isclose(res_analytical, res_autograd).all())

# Calculate gradient of @function with respect to V
res_analytical = grad_V(X, U, D, V).block_until_ready()
res_autograd = autograd_V(X, U, D, V).block_until_ready()
print(jnp.isclose(res_analytical, res_autograd).all())

````

````python
from tqdm.notebook import tqdm
import numpy as np

def times(X,U,D,V):
    '''Count execution time of each function that calculates gradient
    20 times, then calculates and returns mean execution time for each function
    '''
    t1 = %timeit -o -q -n20 grad_U(X,U,D,V).block_until_ready()
    t2 = %timeit -o -q -n20 autograd_U(X,U,D,V).block_until_ready()
    t3 = %timeit -o -q -n20 grad_V(X,U,D,V).block_until_ready()
    t4 = %timeit -o -q -n20 autograd_V(X,U,D,V).block_until_ready()
    t1 = sum(t1.all_runs)/t1.repeat
    t2 = sum(t2.all_runs)/t2.repeat
    t3 = sum(t3.all_runs)/t3.repeat
    t4 = sum(t4.all_runs)/t4.repeat

    return t1,t2,t3,t4

def generate_matrices(n,k):
    '''Generates all the matrices that we need'''
    X = generate(int(n),int(n))
    U = generate(int(n),int(k))
    D = jax.numpy.diag(jax.random.normal(jax.random.PRNGKey(0),(1,int(k)))[0,:])
    V = generate(int(k),int(n))
    return X,U,D,V

c = 100
t1_n, t2_n, t3_n,t4_n = [],[],[],[]
t1_k, t2_k, t3_k,t4_k = [],[],[],[]

for n in tqdm(jnp.linspace(100,10000,20)):
    X, U, D, V = generate_matrices(n,c)
    t1,t2,t3,t4 = times(X, U, D, V)
    t1_n.append(t1)
    t2_n.append(t2)
    t3_n.append(t3)
    t4_n.append(t4)

times_for_k = []
for k in tqdm(jnp.linspace(100,10000,20)):
    X, U, D, V = generate_matrices(c,k)
    t1,t2,t3,t4 = times(X, U, D, V)
    t1_k.append(t1)
    t2_k.append(t2)
    t3_k.append(t3)
    t4_k.append(t4)
````

# Problem 3. Compression of the fully-connected layers in neural network with simple architecture (30 pts)

The main task in this problem is to study how the compression of fully-connected layers affects the test accuracy.
Any fully-connected layer is represented as linear function $AX + B$, where $X$ is input matrix and $A, B$ are trainable matrices. Matrices $A$ in every layer are going to be compressed.

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

````python
def get_accuracy(model, test_loader):
    '''this function is copy of @test, but it does not print the result, but returns it'''
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)
    
    return correct/len(test_loader.dataset)
````

````python
def svd_decompose_layer(layer, compression_rank):
    U,s,V = torch.linalg.svd(layer.weight.data.T)                               # Do SVD decomposition of layer
    b = layer.bias.data                                                         # get bias

    U_r, s_r, V_r  = U[:,:compression_rank], s[:compression_rank], V[:compression_rank,:]   # apply low-rank approximation

    # Make approximated matrices U,s,V and bias learnable parameters
    U_layer = torch.nn.Parameter(U_r, requires_grad = True)
    s_layer = torch.nn.Parameter(s_r, requires_grad = True)
    V_layer = torch.nn.Parameter(V_r, requires_grad = True)
    bias = torch.nn.Parameter(b, requires_grad = True)

    return U_layer, s_layer, V_layer, bias   
    
class SVD_linear(nn.Module):
    '''Linear layer, decomposed by SVD'''

    def __init__(self, base_layer, compression_rank):
        super().__init__()
        self.compression_rank = compression_rank
        self.U, self.s, self.V, self.bias = svd_decompose_layer(base_layer, compression_rank)
        self.shape = [self.U.shape[0] , self.V.shape[1]] # save shape for printing

    def forward(self, x):
        A = self.U @ torch.diag(self.s) @ self.V # construct weight matrix from U s V 
        x = x @ A + self.bias   # linear layer simulation
        return x

    def extra_repr(self):
        # Function for printing the model. Call object to see it
        return '(SVD_linear): Matrix_size={}, compression_rank={}'.format(
            self.shape, self.compression_rank
        )

class SVD_ResidualBlock(nn.Module):
    '''Residual block with decomposed layers by SVD'''

    def __init__(self, base_block,compression_rank):
        super().__init__()

        self.activation = nn.ReLU()
        self.shape = list(base_block.layer1.weight.data.shape) # for printing
        self.compression_rank = compression_rank

        # two linear layers of resblock
        self.layer1 = SVD_linear(base_block.layer1, compression_rank)
        self.layer2 = SVD_linear(base_block.layer2, compression_rank)


    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out 

    def extra_repr(self):
        # Function for printing the model. Call object to see it
        return '(SVD_layer): Matrix_size={}, compression_rank={}'.format(
            self.shape, self.compression_rank
        )


class SVD_Net(nn.Module):
    def __init__(self, base_net, compression_rank):
        '''NN with decomposed layers'''
        super().__init__()

        self.compression_rank = compression_rank
        self.fc1 = SVD_linear(base_net.fc1, compression_rank)        
        self.res_block_1 = SVD_ResidualBlock(base_net.res1, compression_rank)
        self.res_block_2 = SVD_ResidualBlock(base_net.res2, compression_rank)
        self.fc2 = nn.Linear(1000, 10) # last layer is usual Linear layer
        self.ReLU = torch.nn.ReLU()

        # copy params to linear layer from original net
        with torch.no_grad():
            self.fc2.weight.copy_(base_net.fc2.weight.data)
            self.fc2.bias.copy_(base_net.fc2.bias.data)

    def forward(self, x):
        x = self.fc1(x.view(-1,3*32*32))
        x = self.ReLU(x)
        x = self.res_block_1(x)
        x = self.res_block_2(x)
        x = self.fc2(x)       

        return F.log_softmax(x, dim=1)
````

Important thing: Not only linear layer from res blocks were decomposed, also  the very first linear layer, because it has 1000x1000 parameters, which is huge, and we can apply low rank approximation to it.
